```python
import pandas as pd
```

```python
meas = pd.read_csv('Resources/hawaii_measurements.csv')
print(meas.shape)
print(meas.head())
print(meas.dtypes)
print(meas.describe())
print(meas.prcp.unique())
print(meas.tobs.unique())
```

```python
meas_clean = meas.dropna(axis=0, how='any')
print(meas_clean.shape)
```

```python
stn = pd.read_csv('Resources/hawaii_stations.csv')
print(stn)
```

```python
meas_clean.to_csv('clean_measurements.csv', index = False)
stn.to_csv('clean_stations.csv', index = False)
```
